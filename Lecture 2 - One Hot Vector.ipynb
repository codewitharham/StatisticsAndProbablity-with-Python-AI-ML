<h2> Example 1 - Demo </h2>

```python
Animals = ["Cat", "Cat", "Dog", "Bird", "Dog", "Cat", "Bird", "Bird", "Dog", "Cat"]
```

```python
Animals
```

```python
# Making DataFrame using PANDA
# A pandas DataFrame is a primary two-dimensional, size-mutable, and labeled data structure in the Python programming language 
# that is used for data manipulation and analysis
# A Dictionary containing finite numbers of arrays
```

```python
data = {
    "categorical": Animals
}
```

```python
animalDataFrame = pd.DataFrame(data)
```

```python
animalDataFrame
```

```python
animalDataFrame.values
```

```python
# DUMMIES is used to create column coresponding to unique values -- by default consists bool vals

df_dummies = pd.get_dummies(animalDataFrame['categorical'], prefix="category")
```

```python
df_dummies
```

```python
df_dummies = pd.get_dummies(animalDataFrame['categorical'], prefix="category", dtype=int)
```

```python
df_dummies
```

```python
df_dummies.values
```
