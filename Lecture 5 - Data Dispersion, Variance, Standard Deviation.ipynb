## 1. What is Dispersion?
Dispersion (or "spread") tells us about the **diversity** and **reliability** of our data. 

### Why do we care? 
Imagine two weather sensors:
- **Sensor A** stays between $24^{\circ}C$ and $26^{\circ}C$ all day.
- **Sensor B** swings between $10^{\circ}C$ and $40^{\circ}C$ all day.

Both have the same **Mean ($25^{\circ}C$)**, but Sensor A is stable, while Sensor B is highly dispersed.

## 2. Measuring the Spread: Variance & Standard Deviation

To measure dispersion mathematically, we look at how far each point is from the Mean ($\bar{x}$).

### A. Variance ($s^2$ or $\sigma^2$)
Variance is the average of the **squared differences** from the Mean.
- **Formula:** $s^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1}$
- **Why Square it?** If we didn't square the differences, the positive and negative distances would cancel each other out, resulting in zero. Squaring makes everything positive.

### B. Standard Deviation ($s$ or $\sigma$)
Standard Deviation is the square root of the Variance. 
- **Formula:** $s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}}$
- **The "Real World" Metric:** Because variance squares the numbers (turning "cm" into "$cm^2$"), we take the square root to bring the measurement back to the **original units** of our data.






## 3. The "Shape" of Data: Normal vs. Uniform

How data is dispersed depends on its **Distribution**.

### ðŸ§ª Normal Distribution (The Bell Curve)
- Most values cluster near the Mean.
- Values become rarer as you move further away.
- **The 68-95-99.7 Rule:** - $68\%$ of data is within $1$ standard deviation ($\pm 1\sigma$).
    - $95\%$ of data is within $2$ standard deviations ($\pm 2\sigma$).
    - $99.7\%$ of data is within $3$ standard deviations ($\pm 3\sigma$).



### ðŸŸ¦ Uniform Distribution (The Plateau)
- Every value in a range is **equally likely**.
- There is no "peak" or "middle" preference. 
- Example: Rolling a fair die (you are just as likely to get a $1$ as a $6$).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python


# 1. Generate Normal Data (Bell Curve)
# Mean = 50, Std Dev = 10, 1000 samples
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# 2. Generate Uniform Data (Flat)
# Range = 20 to 80, 1000 samples
uniform_data = np.random.uniform(low=20, high=80, size=1000)

# 3. Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot Normal
ax1.hist(normal_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax1.axvline(np.mean(normal_data), color='red', linestyle='dashed', label='Mean')
ax1.set_title(f"Normal Distribution\nMean: {np.mean(normal_data):.2f}, Std Dev: {np.std(normal_data):.2f}")
ax1.legend()

# Plot Uniform
ax2.hist(uniform_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
ax2.axvline(np.mean(uniform_data), color='red', linestyle='dashed', label='Mean')
ax2.set_title(f"Uniform Distribution\nMean: {np.mean(uniform_data):.2f}, Std Dev: {np.std(uniform_data):.2f}")
ax2.legend()

plt.tight_layout()
plt.show()
```

```python

# Settings
sns.set_style("white")
num_samples = 10000

# 1. Generate Uniform Data (Every value between 0 and 10 is equally likely)
uniform_vals = np.random.uniform(low=0.0, high=10.0, size=num_samples)

# 2. Generate Normal Data (Values cluster around the center 'loc')
# loc = mean, scale = standard deviation
normal_vals = np.random.normal(loc=5.0, scale=2.0, size=num_samples)

# Comparison Table
print(f"{'Metric':<20} | {'Uniform':<10} | {'Normal':<10}")
print("-" * 45)
print(f"{'Mean':<20} | {np.mean(uniform_vals):<10.2f} | {np.mean(normal_vals):<10.2f}")
print(f"{'Variance':<20} | {np.var(uniform_vals):<10.2f} | {np.var(normal_vals):<10.2f}")
print(f"{'Std Deviation':<20} | {np.std(uniform_vals):<10.2f} | {np.std(normal_vals):<10.2f}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot Uniform
sns.histplot(uniform_vals, bins=20, ax=ax1, color='skyblue', kde=False)
ax1.set_title("Uniform Distribution (Flat Spread)")
ax1.set_xlabel("Value")

# Plot Normal
sns.histplot(normal_vals, bins=20, ax=ax2, color='salmon', kde=True)
ax2.set_title("Normal Distribution (Bell Curve)")
ax2.set_xlabel("Value")

plt.show()
```

```python
# Practical Outlier Check
mean = np.mean(normal_vals)
sd = np.std(normal_vals)

upper_limit = mean + (3 * sd)
lower_limit = mean - (3 * sd)

print(f"Typical values for this data are between {lower_limit:.2f} and {upper_limit:.2f}")
```
