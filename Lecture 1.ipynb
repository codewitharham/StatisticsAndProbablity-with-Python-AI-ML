```python
plt.plot(demoList)
plt.show()
```

```python
# CREATE FIGURE:
plt.figure(figsize=(10,8))

# add a line renderer with legend and line thickness
plt.plot(demoList)

# making the figure readable--- very important from STAT desc
plt.title("First demo Plot")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.grid()

plt.show()
```

<h1></b> SCATTER PLOT : </h1>

```python
# CREATE FIGURE:
plt.figure(figsize=(10,10))

# add a line renderer with legend and line thickness
mileageDataList = Source["mileage"]
mileageDataListRange = range(mileageDataList.size)
plt.scatter(mileageDataListRange, mileageDataList, marker='x', s = 0.3)

# making the figure readable--- very important from STAT desc
plt.title("Mileage of Toyota's Car")
plt.xlabel("ID's Of Cars")
plt.ylabel("Mileage")
# plt.grid()

plt.show()
```
